```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("covid_19_data.csv")
df.head()
```

```python
df.groupby('ObservationDate').sum(numeric_only=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
```

```python
df.head()
```

```python
df.drop(['SNo','Last Update','Province/State','Country/Region'],axis=1,inplace=True)
df.head()
```

```python
df.groupby('ObservationDate').sum()
```

```python
daily_observation = df.groupby('ObservationDate').sum()
daily_observation.head(3)
```

```python
daily_observation.reset_index(inplace=True)
```

```python
daily_observation['which_day']=(daily_observation['ObservationDate'] - daily_observation['ObservationDate'].min()).dt.days + 1
```

```python
daily_observation.head(5)
```

```python
daily_observation['Infected'] = daily_observation['Confirmed']-daily_observation['Deaths']-daily_observation['Recovered']
daily_observation.head(5)
```

```python
daily_observation.info()
```

```python
#Confirmed
sns.set({'figure.figsize':(14,8)})
sns.regplot(x ='which_day',y='Confirmed', data = daily_observation)
plt.show()
```

```python
#Infected
sns.set({'figure.figsize':(14,8)})
sns.regplot(x ='which_day',y='Infected', data = daily_observation)
plt.show()
```

```python
#Confirmed
sns.set({'figure.figsize':(14,8)})
sns.lineplot(x ='which_day',y='Confirmed', data = daily_observation)
plt.show()
```

```python
#Infected
sns.set({'figure.figsize':(14,8)})
sns.regplot(x ='which_day',y='Deaths', data = daily_observation)
plt.show()
```

```python
#Infected
sns.set({'figure.figsize':(14,8)})
sns.lineplot(x ='which_day',y='Infected', data = daily_observation)
plt.show()
```

```python
#Infected Predict
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
daily_observation.head(5)
```

```python
#target and feature
```

```python
X = daily_observation['which_day'].values.reshape(-1,1)
y = daily_observation['Infected'].values.reshape(-1,1)
```

```python
#test and train
```

```python
model = LinearRegression()
model.fit(X,y)
model.coef_, model.intercept_    #kx+b
model.score(X,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.2,shuffle=False)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
model.score(X_test,y_test)
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
print("R-Squared: ",r2_score(y_test,y_pred))
print("Adjusted R^2: ",1-(1-r2_score(y_test,y_pred))*(n-1)/(n-p-1))
print("Mean Absolute Error: ", mean_absolute_error(y_test,y_pred))
print("Mean Squared Error: ", mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error: ", np.sqrt(mean_squared_error(y_test, y_pred))) 
```
